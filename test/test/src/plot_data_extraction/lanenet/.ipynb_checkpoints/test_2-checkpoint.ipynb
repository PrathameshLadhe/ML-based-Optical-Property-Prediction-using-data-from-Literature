```python
from PIL import Image
import torch
import os
import numpy as np
import glob
import matplotlib.pyplot as plt
from torchvision import transforms
from tqdm import tqdm

from src.models.model import LaneNet
from src.models.loss import DiscriminativeLoss
import cv2

import torch.nn.functional as F
from sklearn.cluster import MeanShift
import matplotlib.colors as mcolors
```

```python
device = torch.device("cuda:1")
```

```python
color_list = mcolors.BASE_COLORS
```

```python
seg_model = LaneNet(cnn_type="unet", embed_dim=4)
seg_model.load_state_dict(torch.load("./checkpoints/simple_seg/plot_digitizer_30.ckpt"))
seg_model.to(device)
seg_model.eval()
```

```python
imglist = glob.glob("/home/weixin/Documents/GitProjects/plot_digitizer/data/val_tmp/*.png")
len(imglist)
```

```python
img = Image.open(imglist[3]).convert("RGB")
plt.imshow(img)
```

```python
img_ = img.crop([71,50,582,435])
w,h = img_.size
```

```python
img_ = img_.resize((w//32*32,h//32*32), resample=Image.NEAREST)
plt.imshow(img_)
img_.size
```

```python
img_np = 255-np.array(img_)
img_np.shape

c = 100

plt.figure(figsize=(20,10))
plt.subplot(1,3,1)
plt.plot(img_np[:,c,0], c=(0.1,0.1,0.1))
plt.subplot(1,3,2)
plt.plot(img_np[:,c,1], c=(0,0.,0.))
plt.subplot(1,3,3)
plt.plot(img_np[:,c,2])
```

```python
img_tensor = transforms.ToTensor()(img_).unsqueeze(0)
img_tensor.shape
```

```python
seg_model.eval()
with torch.no_grad():
    seg_img = seg_model(img_tensor.to(device))[0]
```

```python
seg_mask = torch.max(seg_img, dim=1)[1][0]
seg_mask.shape
```

```python
plt.imshow(seg_mask.data.cpu().numpy(), cmap="gray")
```
