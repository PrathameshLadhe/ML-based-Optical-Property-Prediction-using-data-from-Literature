```python
import requests
import xml.etree.ElementTree as ET
```

```python
MY_API_KEY = ""
```

```python
doilist = []
name = 5404
j = 0
```

```python
import csv

with open('E:/papers/scopus.csv', encoding="utf8") as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    for item in csv_reader:
            print(item[12])
            doilist.append(str(item[12]))
```

```python
print(len(doilist))
```

```python
print(doilist)
```

```python
for doi in doilist:
    url = ("https://api.elsevier.com/content/article/doi/" + doi + "?apiKey=&httpAccept=text%2Fxml")
    resp = requests.get(url,headers={'Accept':'application/xml'})
    if resp.status_code == 200:
        print(resp.status_code,name,j)
        filename = ("E:/papers/" + str(name) + ".xml" )
        with open(filename,'wb') as f:
            for chunk in resp.iter_content(2048):
                f.write(chunk)
            f.close()
        name = name + 1
    j = j+1


    
```
