```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.tri as tri


# Create a triangulation from x, y values
triang = tri.Triangulation(x, y)

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surf = ax.plot_trisurf(x, y, z, cmap='plasma', antialiased=False)

# Add contour lines
contour = ax.tricontour(triang, z, levels=10, colors='k', linestyles='solid')

# Set labels for x, y, z axes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Add a colorbar
fig.colorbar(surf, ax=ax, label='Z')

ax.view_init(elev=15, azim=110)

# Show the plot
plt.show()
```
