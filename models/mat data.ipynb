```python
import pandas as pd
import json
```

```python
with open(r'') as f:
    data = json.load(f)
```

```python
data[1]['id']
```

```python
columns = []
listf = ['id','img','leg', 'unit', 'dep_temp', 'anneal_temp','anneal_time','thickness','temp','pressure', 'x_val','y_val']
```

```python
for record in data:
    for element in record:
        if element not in columns:
            columns.append(element)
```

```python
df = pd.DataFrame(columns=columns)

print(df)
```

```python
for record in data:
    row = {}
     
    temp_list = record.keys()
    print(temp_list)
    
    #row = {'id' : record['id'], 'img' : record['img'], 'leg' : record['leg']}
    for element in columns:
        if element not in listf and element in temp_list :
            print(record['id'])
            row[element] = record[element]
        else:
            row[element] = 0
    for elem in listf:
        row[elem] = record[elem]    
    df = df.append(row, ignore_index=True)
df.set_index('id', inplace=True)
```
